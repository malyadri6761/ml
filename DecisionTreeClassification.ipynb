```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,recall_score,precision_score,f1_score,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
data = pd.read_csv('/content/drive/MyDrive/Datasets/LogisticRegression/Social_Network_Ads.csv')
```

```python
data.head()
```

```python
data = data.drop(columns=['User ID'],axis=1)
```

```python
data.head()
```

```python
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
scaler = StandardScaler()
data[['Age','EstimatedSalary']] = scaler.fit_transform(data[['Age','EstimatedSalary']])
```

```python
data.head()
```

```python
X = data.drop(columns=['Purchased'],axis=1)
y = data['Purchased']
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=0.2,random_state=True)
```

```python
model = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
model.fit(train_x,train_y)
```

```python
model.predict(train_x)
```

```python
model.predict(test_x)
```

```python
print('accuracy score : ',accuracy_score(test_y,model.predict(test_x)))
print('recall score : ',recall_score(test_y,model.predict(test_x)))
```
