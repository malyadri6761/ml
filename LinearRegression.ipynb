```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error
import numpy as np
```

```python
train = pd.read_csv('/content/drive/MyDrive/Datasets/LinearRegression/train.csv')
test = pd.read_csv('/content/drive/MyDrive/Datasets/LinearRegression/train.csv')
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,6))

ax[0,0].plot(train['x'],train['y'])
ax[0,0].set_title('Train')
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y')

ax[0,1].scatter(train['x'],train['y'],color='red')
ax[0,1].set_title('Train')
ax[0,1].set_xlabel('x')
ax[0,1].set_ylabel('y')

ax[1,0].plot(test['x'],test['y'])
ax[1,0].set_title('Test')
ax[1,0].set_xlabel('x')
ax[1,0].set_ylabel('y')

ax[1,1].scatter(test['x'],test['y'],color='green')
ax[1,1].set_title('Test')
ax[1,1].set_xlabel('x')
ax[1,1].set_ylabel('y')
```

```python
model = LinearRegression()
model.fit(np.array(train['x']).reshape(-1,1),train['y'])
```

```python
model.predict(np.array(test['x']).reshape(-1,1))
```

```python
test=test.dropna()
```

```python
print(r2_score(test['y'],model.predict(np.array(test['x']).reshape(-1,1))))
print(mean_squared_error(test['y'],model.predict(np.array(test['x']).reshape(-1,1))))
print(mean_absolute_error(test['y'],model.predict(np.array(test['x']).reshape(-1,1))))
print(root_mean_squared_error(test['y'],model.predict(np.array(test['x']).reshape(-1,1))))
```
